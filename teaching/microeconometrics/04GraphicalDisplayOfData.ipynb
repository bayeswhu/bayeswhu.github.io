## Graphical display of data

- Graphs visually demonstrate important features of the data. 
- Different types of data require distinct graph formats to bring out these features.

- `Seaborn` is a library for making statistical graphics in Python. 
- It is built on top of `matplotlib` and closely integrated with `pandas` data structures.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_stata('data/mus02psid92m.dta')
```

- For example, if we want to produce a scatterplot of `lnearns` and `hours`

```python
sns.relplot(x='hours', y='lnearns', data=data);
```

- In practice, however, customizing is often desirable. 
- For example, we may want to display the relationship between lnearns and hours by showing both the data scatterplot and the ordinary least-squares (OLS) fitted line on the same graph. 

```python
sns.lmplot(x='hours', y='lnearns', data=data);
```

###  Box-and-whisker plot

- Box-and-whisker plot is a graphical way to display data on a single series. 
- The boxes cover the interquartile range, from the lower quartile to the upper quartile. 
- The whiskers, denoted by horizontal lines, extend to cover most or all the range of the data.
- We want to present separate box plots of hours for each of four education groups

```python
sns.boxplot(x="edcat", y="hours", data=data);
```

###  Histogram

- The probability mass function or density function can be estimated using a histogram

```python
sns.histplot(data['lnearns'].dropna());
```

###  Kernel density plot

- For continuous data taking many values, a better alternative to the histogram is a kernel density plot. 
- This provides a smoother version of the histogram
- If we desire a smoother density, we could use the following code:

```python
from scipy.stats import norm
sns.kdeplot(data['lnearns'].dropna());
```

- The following code instead presents a histogram overlaid by a kernel density estimate.

```python
sns.displot(data['lnearns'].dropna(),kde=True);
```

### Twoway scatterplots and fitted lines

- Scatterplots provide a quick look at the relationship between two variables.
- It can be useful to additionally provide a fi tted curve.
- For example, we may want to provide a scatterplot and fitted quadratic with confidence bands

```python
sns.lmplot(x='hours',y='lnearns',data=data,order=2);
```

## Multiple scatterplots

- Here we produce bivariate scatterplots of lnearns, hours, and age:

```python
g=sns.PairGrid(data,vars=['lnearns','hours','age'])
g.map(plt.scatter);
```
