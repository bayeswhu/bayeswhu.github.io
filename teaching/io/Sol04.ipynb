### Exercise 5

```python
from sympy import divisor_count

currentNum = 2
currentTotalNum = 1

while 1:
    if divisor_count(currentTotalNum) > 400:
        print(currentTotalNum)
        break
    else:
        currentTotalNum += currentNum
        currentNum += 1
```
